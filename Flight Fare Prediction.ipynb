# Preprocessing

```python
# Total_Stops is an ordinal categorical data type we perform LabelEncoder

stops_dict = {'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4}
data['Total_Stops'] = data['Total_Stops'].map(stops_dict)
```

```python
data.Date_of_Journey = pd.to_datetime(data.Date_of_Journey)
```

```python
# Extract day, month and year from date of journey

data['Journey_Month'] = data.Date_of_Journey.dt.month
data['Journey_Day'] = data.Date_of_Journey.dt.day
data['Journey_Year'] = data.Date_of_Journey.dt.year
```

```python
data.Journey_Year.unique()
```

```python
# Data has only 1 year so we can drop the year data

data.drop('Journey_Year',axis=1,inplace=True)
```

```python
# Checking whether range of months is out of bound or not

data['Journey_Month'].unique()
```

```python
# Checking whether range of days is out of bound or not

data['Journey_Day'].unique()
```

```python
data.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
# From departure time we can extract the time when plane leaves the gate

data['Dep_min'] = pd.to_datetime(data['Dep_Time']).dt.minute
data['Dep_hour'] = pd.to_datetime(data['Dep_Time']).dt.hour
```

```python
data.drop('Dep_Time',axis=1,inplace=True)
```

```python
# Checking whether range of hour is out of bound or not

data.Dep_hour.unique()
```

```python
# Checking whether range of minutes is out of bound or not

data.Dep_min.unique()
```

```python
# Extracting the time when plane arrives at its destination

data['Arrival_min'] = pd.to_datetime(data['Arrival_Time']).dt.minute
data['Arrival_hour'] = pd.to_datetime(data['Arrival_Time']).dt.hour
```

```python
data.drop('Arrival_Time',axis=1,inplace=True)
```

```python
# Checking whether range of hour is out of bound or not

data.Arrival_hour.unique()
```

```python
# Checking whether range of minutes is out of bound or not

data.Arrival_min.unique()
```

```python
# Here is duration is only 5 mins which is irrelevant

data[data.Duration=='5m']
```

```python
data.drop(6474,inplace=True)
```

```python
data[data.Duration=='5m']
```

```python
data.index = [i for i in range(len(data))]
```

```python
hours = []
minutes = []

for i in range(len(data)):
    time = data['Duration'][i].split()
    
    hr = ''
    mn = ''
    
    if len(time)==1:
        hr = time[0]
    else:
        hr,mn = time[0],time[1]
        
    if hr:
        hours.append(int(hr.replace('h','')))
    else:
        hours.append(0)
    if mn:
        minutes.append(int(mn.replace('m','')))
    else:
        minutes.append(0)
    
```

```python
data['Duration_hours'] = hours
data['Duration_min'] = minutes
```

```python
data.drop('Duration',inplace =True,axis=1)
```

```python
(data.Additional_Info.value_counts()/len(data))*100
```

```python
# 78% data has addtional info as no info 
# Route and Total_Stops are related to each other

data.drop(['Route','Additional_Info'],axis=1,inplace=True)
```

```python
# Airline, Source and Destination are nominal categorical data we perform OneHotEncoding

airline = data[['Airline']]
airline = pd.get_dummies(airline,drop_first=True)
airline.head()
```

```python
source = data[['Source']]
source = pd.get_dummies(source,drop_first=True)
source.head()
```

```python
destination = data[['Destination']]
destination = pd.get_dummies(destination,drop_first=True)
destination.head()
```

```python
data = pd.concat([data,airline,source,destination],axis=1)
```

```python
data.drop(['Airline','Source','Destination'],inplace=True,axis=1)
```

```python
data.head()
```

# Model Building

```python
# Independent Features
X = data.drop('Price',axis=1)
```

```python
X.head()
```

```python
# Dependent Feature
y = data.Price
```

```python
y.head()
```

```python
# Correraltion between independent and dependent feature

plt.figure(figsize=(24,24))
sns.heatmap(data.corr(),annot=True,cmap='Blues')
plt.show()
```

```python
# Important feature using ExtraTreesRegressor

from sklearn.ensemble import ExtraTreesRegressor
selection = ExtraTreesRegressor()
selection.fit(X, y)
```

```python
selection.feature_importances_
```

```python
plt.figure(figsize = (12,8))
feat_importances = pd.Series(selection.feature_importances_, index=X.columns)
feat_importances.nlargest(20).plot(kind='barh')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 21)
```

```python
from sklearn.ensemble import RandomForestRegressor
reg_rf = RandomForestRegressor()
reg_rf.fit(X_train, y_train)
```

```python
y_pred = reg_rf.predict(X_test)
```

```python
reg_rf.score(X_train, y_train)
```

```python
reg_rf.score(X_test, y_test)
```

```python
from sklearn import metrics

metrics.r2_score(y_test, y_pred)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Randomized Search CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100,stop=1200,num=12)]

# Number of features consider at every split
max_features = ['auto','sqrt']

# Maximum depth of tree
max_depth = np.linspace(start=5,stop=30,num=6)

# Minimum number of samples required to split a node
min_samples_split = [2,5,10,15,100]

# Minimum number of samples required at each leaf node
min_samples_leaf = [1,2,5,10]
```

```python
random_grid = {'n_estimators':n_estimators,
               'max_features':max_features,
               'max_depth':max_depth,
               'min_samples_split':min_samples_split,
               'min_samples_leaf':min_samples_leaf}
```

```python
rf_random = RandomizedSearchCV(reg_rf,param_distributions=random_grid,verbose=2)
```

```python
rf_random.fit(X_train,y_train)
```

```python
rf_random.best_params_
```

```python
prediction = rf_random.predict(X_test)
```

```python
metrics.r2_score(y_test, prediction)
```
